```python
import pandas as pd
import numpy as np
df= pd.read_excel("Sales Data Presentation - Dashboards.xlsx",index_col = 0)
print(df.head())
```

```python
import matplotlib.pyplot as plt

menos_vendidos = df.groupby("ProdName")["Quantity"].sum().nsmallest(5)
plt.figure(figsize=(8,5))
menos_vendidos.plot(kind="bar", color="skyblue", edgecolor="white")
plt.title("5 productos menos vendidos")
plt.ylabel("Cantidad vendida")
plt.xlabel("Producto")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
producto_especifico = df[df["ProdName"] == "Scented Olive Oil Lotion Gift Basket"]["Quantity"]
plt.figure(figsize=(6,4))
plt.boxplot(producto_especifico, vert=False, patch_artist=True, boxprops=dict(facecolor="lightgreen"))
plt.title("Distribución de Quantity - Scented Olive Oil Lotion Gift Basket")
plt.xlabel("Quantity")
plt.show()
```

```python
import matplotlib.pyplot as plt
bath_products = df[df["ProdCategory"]=="Bath products"].groupby("ProdName")["Quantity"].sum()
plt.figure(figsize=(10,5))
plt.plot(bath_products.index, bath_products.values, marker="o", linestyle="-", color="purple")
plt.title("Cantidad de productos vendidos - Categoría Bath Products")
plt.ylabel("Quantity")
plt.xlabel("Producto")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```
